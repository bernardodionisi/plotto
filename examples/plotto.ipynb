# Plotto

```python
from plotto import load_data, mark_plot
from vega_datasets import data
```

```python
# click on legend to select by group
```

## Scatterplot

```python
df = data.cars()

mark_plot(
    data=df, 
    x='Horsepower', 
    y='Miles_per_Gallon', 
    shape_by='Origin',
    opacity=0.75,
    y_title=['Miles', 'per Gallon']
)
```

## Plot parameters

```python
params = load_data.params()

params
```

```python
df = (
    params
    .loc[lambda x: (x['sample_name'] == 'samples = 0') & (x['stratum'] == 0)]
    .assign(cohort=lambda x: x['cohort'].astype(str))
)

mark_plot(
    data=df, 
    x='time', 
    y='param', 
    shape_by='cohort', 
    lines=True,
    configure_axisX={'format': 'c'},
    dashes=[[1,1], [1, 0]],
    shapes=['triangle', 'circle'],
    colors='black',
)
```

```python
mark_plot(
    data=df, 
    x='time', 
    y='param',
    lower_upper=['lower', 'upper'],
    shape_by='cohort', 
    lines=True,
    points=False,
    ebands=True,
    configure_legend_color_by={'symbolSize': 300}
)
```

```python
df = (
    params
    .loc[lambda x: (x['cohort'] == 1904) & (x['time'] == 1901) & (x['stratum'] == 1)]
    .assign(cohort=lambda x: x['cohort'].astype(str))
)
```

```python
mark_plot(
    data=df, 
    x='param', 
    y='sample_name',
    lower_upper=['lower', 'upper'],
    ebars=True,
    eticks=True,
    vertical=False,
    y_title=['Sample', 'Names'],
    vertical_line=0,
    configure_axisY={'domain': True, 'ticks': True},
    configure_axisX={'grid': True},
    configure_vertical_line={'strokeDash': [0, 0]}
)
```
